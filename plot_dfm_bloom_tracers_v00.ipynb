```python
def ratio(a,b,b_min=1e-8):
    return a/b.clip(b_min) * np.where(b<b_min,np.nan, 1.0)
```

```python
frac=0.1
M=grid.smooth_matrix()
def fill(values,iterations=50):
    valid=np.isfinite(values)
    data=np.where(valid,values,0.0)
    weight=np.where(valid,1,0.0)
    for _ in range(iterations):
        data=M.dot(data)
        weight=M.dot(weight)
        data[valid]=values[valid]
        weight[valid]=1.0
    result=np.full(len(values),np.nan)
    valid=weight>1e-4
    result[valid]=data[valid]/weight[valid]
    return result
```

```python
from stompy import memoize

@memoize.memoize(lru=1)
def extract_tracers(tidx,layer,Isat=10.0):
    # Extraction
    # instantaneous release, so age is a given.
    t=ds.time.isel(time=tidx).values
    age_d = (t - ds.time.values[0]) / np.timedelta64(86400,'s')
    
    if layer=='mean':
        snap=ds.isel(time=tidx)
        conc = snap['conc0'].values.mean(axis=1).clip(0,np.inf)
    
        #Imean =  ratio( snap['radc0'].values, snap['agec0'].values, thresh)
        Imean = ratio( snap['radc0'].values, snap['conc0'].values*age_d, thresh)
    
        Imean[np.isnan(Imean)]=0.0 
        
        kLight = (Imean/(Imean + Isat)).mean(axis=1) # happen to know that layer is the second dimension    
    else:
        snap=ds.isel(time=tidx,laydim=layer)
    
        # gather inputs
        # average irradiance accounting for Kd, vertical mixing.
        Imean =  ratio( snap['radc0'].values, snap['agec0'].values, thresh)
        # This could be averaged over the layers instead.
        kLight = Imean/(Imean + Isat)
        conc  = snap['conc0'].values

    kLight = fill(kLight, iterations=120)
    return dict(age_d=age_d, conc=conc, kLight=kLight, Imean=Imean, t=t)  
```

```python
# Panel for P and a panel for N.

def solveNP(thresh=1e-5,
            # Parameters:
            c0 = 50,
            tidx = 20, 
            Isat=10.,
            N0=35, # initial DIN
            Nsat=0.5,
            kprod=0.7, # per day.
            kmort=0.1,
            alpha=0.15, # stoichiometry, uM DIN per ug/l chl production.
            layer='mean'):
    tracers=extract_tracers(tidx=tidx, layer=layer, Isat=Isat)
    kLight = tracers['kLight']
    conc = c0*tracers['conc']
    age_d = tracers['age_d']
    Imean = tracers['Imean']
    t = tracers['t']

    if 0: # Super simplified
        # Very rough calculation up to the point of nutrient limitations
        kDIN = 1.0 # assuming no limitation.
        mu_net = (kprod-kmort)*kLight*kDIN # 1/d
        
        # Solve for the amount of time needed to exhaust nitrogen, or age, whichever is
        # smaller.
        
        # conc*(np.exp(mu_net* t_nutrient_d) - 1) = N0/alpha
        # np.exp(mu_net * t_nutrient_d) - 1 = N0/(alpha * conc)
        # t_nutrient_d = np.log(N0/(alpha*conc) + 1) / mu_net
        
        valid = (conc>0.0) & (mu_net>0)
        t_nutrient_d = np.full(conc.shape,np.inf)
        t_nutrient_d[valid] = np.log(N0/(alpha*conc[valid]) + 1) / mu_net[valid]
        
        t_d=np.minimum( age_d, t_nutrient_d)
        
        # total growth of biomass.
        P = conc * np.exp(mu_net * t_d)
        P[conc<thresh]=conc[conc<thresh]
        N = N0 - alpha * (P - conc)
    if 1: # ODE integration
        # State vector is [ P[0].... P[i], N[0], ... , N[i] ]
        def diff(t,state):
            P,N=state.reshape([2,-1])
            N=N.clip(0) # no evidence this is really necessary
            P=P.clip(0) # likewise.
            kDIN=N/(N+Nsat)
            dgrossP = kprod*kLight*kDIN*P
            dnetP = -kmort*P + dgrossP
            dN = -alpha*dgrossP
            mu_net = kprod*kLight*kDIN - kmort
            return np.r_[dnetP,dN]
        IC=np.r_[conc, N0*np.ones_like(conc)]
        # Looks like the shorter the steps are the lower the final answer.
        # Is that a bug?  hmax=0.01 => 
        # constant kDIN gives 11.92, 33.11
        # true kDIN gives 10.44, N down to 33.399 
    
        # odeint was buggy.
        #result = odeint(diff, IC, [0,age_d], hmax=0.1)
        bunch = solve_ivp(diff, y0=IC, t_span=[0,age_d])
        result=bunch.y[:,-1]
        P,N = result.reshape([2,-1])
    ds=xr.Dataset()
    ds['N'] = ('cell',), N
    ds['P'] = ('cell',), P
    ds['time'] = (),t
    return ds
```
